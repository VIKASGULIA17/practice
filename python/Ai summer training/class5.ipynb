```python
x=range(10,40)
y=np.linspace(0,80,30)
```

```python
x2=[1,2,3,4]
y2=[[4,2,7,4],
   [12,5,3,7],
   [8,23,23,35]]
```

```python
x3=['a','b','c','d']
y3=[32,45,65,13]
```

```python
x7=['a','b','c','d']
y7=[32,45,65,13,43,45,234,267,46,75,23,345,7,345]
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(18,18))

# Line Plot
ax1 = plt.subplot(3,3,1)
ax1.plot(x, y, linestyle='--', marker='o', label='Line data')
for i in range(len(x)):
    ax1.text(x[i], y[i], str(round(y[i],1)), fontsize=8)
ax1.set_title('Line Plot')
ax1.set_xlabel('x')
ax1.set_ylabel('y')
ax1.legend()

# Stack Plot
ax2 = plt.subplot(3,3,2)
ax2.stackplot(x2, y2, labels=['A','B','C','D'])
ax2.set_title('Stack Plot')
ax2.set_xlabel('x')
ax2.set_ylabel('y')
ax2.legend()

# Bar Chart
ax3 = plt.subplot(3,3,3)
bars = ax3.bar(x3, y3, color=['g', 'b','r', 'k'], label=['A','B','C','D'])
ax3.set_title('Bar Chart')
ax3.set_xlabel('x')
ax3.set_ylabel('y')
ax3.bar_label(bars)
ax3.legend()

# Horizontal Bar Chart
ax4 = plt.subplot(3,3,4)
bars_h = ax4.barh(x3, y3, color=['g', 'b','r', 'k'], label=['A','B','C','D'])
ax4.set_title('Horizontal Bar Chart')
ax4.set_xlabel('x')
ax4.set_ylabel('y')
ax4.bar_label(bars_h)
ax4.legend()

# Scatter Plot
ax5 = plt.subplot(3,3,5)
ax5.scatter(x, y, alpha=0.3, label='Points')
for i in range(len(x)):
    ax5.text(x[i], y[i], str(round(y[i],1)), fontsize=8)
ax5.set_title('Scatter Plot')
ax5.set_xlabel('x')
ax5.set_ylabel('y')
ax5.legend()

# Box Plot
ax6 = plt.subplot(3,3,6)
ax6.boxplot(x6)
ax6.set_title('Box Plot')
ax6.set_xlabel('x')
ax6.set_ylabel('y')

# Histogram
ax7 = plt.subplot(3,3,7)
ax7.hist(y7, bins=5, label='Histogram data')

ax7.set_title('Histogram')
ax7.set_xlabel('x')
ax7.set_ylabel('y')
ax7.legend()

# Violin Plot
ax8 = plt.subplot(3,3,8)
ax8.violinplot(y7)
ax8.set_title('Violin Plot')
ax8.set_xlabel('x')
ax8.set_ylabel('y')

# Pie Chart
ax9 = plt.subplot(3,3,9)
ax9.pie(y3, labels=x3, autopct='%1.1f%%')
ax9.set_title('Pie Chart')

plt.show()
```
