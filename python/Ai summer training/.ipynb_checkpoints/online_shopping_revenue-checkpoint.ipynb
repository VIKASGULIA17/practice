```python
# Create a copy for preprocessing
df_processed = df.copy()
```

```python
# Encode Month
le = LabelEncoder()
df_processed['Month_encoded'] = le.fit_transform(df_processed['Month'])
df_processed.drop(columns=['Month'],inplace=True)
```

```python
# Encode Month
le = LabelEncoder()
df_processed['VisitorType_encoded'] = le.fit_transform(df_processed['VisitorType'])
df_processed.drop(columns=['VisitorType'],inplace=True)
```

```python
def handling_outliers(df, col):
    q1 = df[col].quantile(0.25)
    q3 = df[col].quantile(0.75)
    iqr = q3 - q1
    lower = q1 - 1.5 * iqr
    upper = q3 + 1.5 * iqr
    df = df[(df[col] >= lower) & (df[col] <= upper)]
    return df
```

```python
df.shape
```

```python
# Select features for modeling
feature_columns = [
    'Administrative', 'Administrative_Duration', 'Informational', 'Informational_Duration',
    'ProductRelated', 'ProductRelated_Duration', 'BounceRates', 'ExitRates', 
    'PageValues', 'SpecialDay', 'Month_encoded', 'OperatingSystems', 'Browser',
    'Region', 'TrafficType', 'VisitorType_encoded', 'Weekend'
]

x = df_processed[feature_columns]
y = df_processed['Revenue']
```

```python
# Converting boolean columns to integers
df_processed['Weekend'] = df_processed['Weekend'].astype(int)
df_processed['Revenue'] = df_processed['Revenue'].astype(int)
outlier_columns = ['Administrative_Duration', 'Informational_Duration', 'ProductRelated_Duration']
for col in outlier_columns:
    combined_df = pd.concat([x, y], axis=1)
    combined_df = handling_outliers(combined_df, col)
    x = combined_df[feature_columns]
    y = combined_df['Revenue']
df.shape
```

# TRAIN-TEST SPLIT

```python
# spliting data into train test
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
# Scale the features
scaler = StandardScaler()
x_train_scaled = scaler.fit_transform(x_train)
x_test_scaled = scaler.transform(x_test)
```

# LOGISTIC REGRESSION MODEL

```python
# creating logistic regresion class
lg=LogisticRegression()
lg.fit(x_train_scaled,y_train)
```

```python
# Making predictions
y_pred = lg.predict(x_test_scaled)
```

```python
accuracy = accuracy_score(y_test, y_pred)
accuracy
```

# Decision tree

```python
# creating model of decision tree

from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
dtc = DecisionTreeClassifier()
dtc.fit(x_train,y_train)
y_pred2 = dtc.predict(x_test)
score2 = accuracy_score(y_test,y_pred2)
print(score2)
```

# Random forest classifier

```python
# creating model of Random forest classifier

rfc = RandomForestClassifier()
rfc.fit(x_train,y_train)
y_pred3 = rfc.predict(x_test)
score3 = accuracy_score(y_test,y_pred3)
print(score3)
```

# NAIVE BAYES

```python
# creating model of Naive bayes

gnb=GaussianNB()
gnb.fit(x_train,y_train)
y_pred4=gnb.predict(x_test)
score4 = accuracy_score(y_test,y_pred4)
score4
```

# SVC

```python
# creating model of svc
svc=SVC(kernel='linear')
svc.fit(x_train_scaled,y_train)
y_pred5=svc.predict(x_test_scaled)
score5 = accuracy_score(y_test,y_pred5)
score5
```

# KNN

```python
# creating model of knn

knn=KNeighborsClassifier()
knn.fit(x_train,y_train)
y_pred6=knn.predict(x_test)
score6 = accuracy_score(y_test,y_pred6)
score6
```
